```python
import re

def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # remove urls
    text = re.sub(r"http\S+|www\S+", "", text)

    # remove emails
    text = re.sub(r"\S+@\S+", "", text)

    # remove numbers
    text = re.sub(r"\d+", "", text)

    # remove repeated characters (soooo -> soo)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)

    # convert to lowercase
    text = text.lower()

    # remove special characters
    text = re.sub(r"[^a-z\s]", " ", text)

    # remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    # remove short tokens
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            cleaned_tokens.append(word)

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for s in sentences:
    tokens, cleaned = preprocess_text(s)

    print("Original:", s)
    print("Tokens:", tokens)
    print("Clean Sentence:", cleaned)
    print("-"*50)

    results.append(tokens)
```

```python
import numpy as np

for i, tokens in enumerate(results):

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = np.mean([len(t) for t in tokens])
    else:
        avg_length = 0

    print(f"Sentence {i+1}")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length,2))
    print("-"*40)
```

```python
from collections import Counter

all_tokens = []

for tokens in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent words")
print(counter.most_common(10))

print("\nTop 5 least frequent words")
least_common = counter.most_common()[:-6:-1]
print(least_common)
```

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        tokens, cleaned = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
pipeline_output = full_pipeline(sentences)

print(pipeline_output)
```
